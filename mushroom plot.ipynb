```python
import matplotlib.pyplot as plt
import pandas as pd
msroom = pd.read_csv('G:\My Study\Springboard\Projects\Capstone1\mushrooms.csv')
```

```python
import seaborn as sns
sns.countplot(x='cap-shape',hue='class',data=msroom)
plt.show()
```

```python
for column in msroom.columns[1:]:
    sns.countplot(x=column, hue='class', data = msroom)
    plt.show()
```

Due to the fact that all variables are catagorical, only bar plots are made. 
It is not clear to see the correlations between poisonous/edible and other variables, but in some variables, one property alone can indicated the mushrooms' edability, such as the 'cap-shape', all the 'sunken' shape mushrooms are edible, and all the 'conical' shape mushrooms are poisonous.
