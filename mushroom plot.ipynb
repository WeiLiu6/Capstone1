```python
import matplotlib.pyplot as plt
import pandas as pd
msroom = pd.read_csv('G:\My Study\Springboard\Projects\Capstone1\Data\mushrooms.csv')
```

```python
msroom
```

```python
import seaborn as sns
a = sns.countplot(x='cap-shape',hue='class',data=msroom)

leg = a.get_legend()
leg.set_title("Mushroom Class")
labs = leg.texts
labs[0].set_text("Poisonous")
labs[1].set_text("Edible")

plt.show()
```

```python
for column in msroom.columns[1:]:
    a = sns.countplot(x=column, hue='class', data = msroom)
    leg = a.get_legend()
    leg.set_title("Mushroom Class")
    labs = leg.texts
    labs[0].set_text("Poisonous")
    labs[1].set_text("Edible")

    plt.show()
```

Due to the fact that all variables are catagorical, only bar plots are made. 
It is not clear to see the correlations between poisonous/edible and other variables, but in some variables, one property alone can indicated the mushrooms' edability, such as the 'cap-shape', all the 'sunken' shape mushrooms are edible, and all the 'conical' shape mushrooms are poisonous.
