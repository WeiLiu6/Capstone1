```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
msroom = pd.read_csv('G:\My Study\Springboard\Projects\Capstone1\Data\mushrooms.csv')
msroom.head()
```

```python
#Calculate the correlation between every variable and 'class'
from scipy.stats import chi2_contingency
chi = {}
for column in msroom.columns[1:]:
    left = pd.DataFrame(msroom[msroom['class']=='p'][column].value_counts())
    right = pd.DataFrame(msroom[msroom['class']=='e'][column].value_counts())
    joined = left.join(right,lsuffix='_p',rsuffix='_e',how='outer')
    joined[np.isnan(joined)] = 0
    #Using the chi2_contingency() to calculate the chi-square value and p value
    chi[column] = chi2_contingency(joined)[0:2]
chi
```

It can be seen from the result, the variable 'odor' and 'spore-print-color' have the largest chi square value(7659.7 and 4602.0), meaning that these two variables have the largest correlation between edibility. And correlation between the variable 'veil-type' and edibility('class') is 0, this variable can be ignored when try to decide a mushroom's edibility.
